#### Prostorne krivulje

Skup od $\infty^1$ neprekinuto povezanih točaka prostora, koje ne leže u jednoj ravnini, nazivamo __prostornom krivuljom__. Loksodroma je primjer jedne prostorne krivulje koju sam već pokazala. Sada ćemo vidjeti još neke primjere prostornih krivulja. 

###### Prostorna kubna parabola

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

t = np.linspace(-3, 3, 100)
x = t
y = t**2
z = t**3

ax.plot(x, y, z, label='Prostorna kubna parabola')
ax.legend()

plt.show()

```

###### Obična cilindrična spirala

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import matplotlib as mpl
import numpy as np
import math

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

t = np.linspace(-2, 14, 200)
z = t
y = np.sin(t)
x = np.cos(t)
ax.plot(x, y, z, label='Cilindrična spirala')
ax.legend()

plt.show()
```

###### Vivijanijev prozor

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import axes3d
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
r = 3
t = np.linspace(-4*np.pi, 4*np.pi, 100)
x = r*(1 + np.cos(t))
y = r*np.sin(t)
z = r*2*np.sin(t/2)
ax.plot(x, y, z, label='Vivijanijev prozor')
ax.legend()

plt.show()
```
